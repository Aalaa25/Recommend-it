# Import Libraries

```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, SVD,KNNWithMeans
from surprise.prediction_algorithms.matrix_factorization import SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import mse, mae, rmse
from surprise import accuracy
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

# SVD Model

```python
reader = Reader(rating_scale=(0.5, 5))
```

```python
data =Dataset.load_from_df(rating_df[['userId', 'movieId', 'rating']], reader)
data
```

```python
svd_model=SVD(biased=False)

```

```python
trainset, testset = train_test_split(data, test_size=0.25, random_state=123)
```

```python
trainset.n_items, trainset.n_users, trainset.n_ratings
```

```python
svd_model.fit(trainset)
```

```python
import pickle
with open('svd_model.pkl', 'wb') as f:
    pickle.dump(svd_model, f)
```

```python
with open('svd_model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
preditions=model.test(testset)
```

```python
mae(preditions), mse(preditions), rmse(preditions)
```

## Data preprocessing

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity

movie_merge['genres'] = movie_merge['genres'].fillna('(no genres listed)')
movie_merge['genres'] = movie_merge['genres'].apply(lambda x: x.split('|') if isinstance(x, str) else [])

mlb = MultiLabelBinarizer()
genre_onehot = pd.DataFrame(
    mlb.fit_transform(movie_merge['genres']),
    columns=mlb.classes_,
    index=movie_merge.index
)
```

# similarity based on genre and rating

```python
genre_onehot['movieId'] = movie_merge['movieId']
genre_onehot['rating']=movie_merge['rating']

genre_features = genre_onehot.drop_duplicates('movieId').set_index('movieId').drop(columns='movieId', errors='ignore')

similarity_matrix = cosine_similarity(genre_features)

similarity_df = pd.DataFrame(similarity_matrix, index=genre_features.index, columns=genre_features.index)

```
